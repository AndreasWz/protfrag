```python
import pandas as pd
import ijson
from tqdm import tqdm
```

```python
with open("./real_fragments30.txt", "r") as f:
    fragment_ids = set([line[:-1] for line in f.readlines()])

len(fragment_ids)
```

```python
data = []

with open("../../data/uniprotkb_reviewed_true_2025_11_07.json", "r") as f:
    for record in tqdm(ijson.items(f, "results.item"), desc="Processing proteins"):
        acc_id = record["primaryAccession"]
        sequence = record["sequence"]["value"]
        if acc_id in fragment_ids:
            data.append((acc_id, sequence))
```

```python
len(data)
```

```python
data[0]
```

```python
df = pd.DataFrame(data, columns=["source_accession", "sequence"])
df.head()
```

```python
df["fragment_length"] = df["sequence"].apply(lambda seq: len(seq))
df["is_fragment"] = True
```

```python
df.to_csv("./real_fragments30.csv", index=False)
```
