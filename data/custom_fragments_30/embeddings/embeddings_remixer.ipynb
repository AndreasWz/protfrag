```python
import pandas as pd
import pickle
import random
import torch
```

```python
with open("./final_embeddings_real_fragments.pkl", "rb") as p:
    fragments = pickle.load(p)

fragments
```

```python
unique = []

for item in fragments:
    is_duplicate = False
    for u in unique:
        if item[0] == u[0] and item[1] == u[1] and torch.equal(item[2], u[2]):
            is_duplicate = True
            break

    if not is_duplicate:
        unique.append(item)

len(unique)
```

```python
fragments = unique
```

```python
with open("./final_embeddings_train.pkl", "rb") as p:
    train = pickle.load(p)
```

```python
len(train)
```

```python
def deduplicate(data):
    unique = []
    for item in data:
        is_duplicate = False
        for u in unique:
            if item[0] == u[0] and item[1] == u[1] and item[2] == u[2] and torch.equal(item[3], u[3]):
                is_duplicate = True
                break

        if not is_duplicate:
            unique.append(item)

    return unique
```

```python
train = deduplicate(train)
```
