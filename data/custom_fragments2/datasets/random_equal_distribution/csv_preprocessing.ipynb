```python
import pandas as pd

df = pd.read_csv("run_random_equal_val.csv")

# 1) create frag_idx 
df["frag_idx"] = df.groupby("source_accession").cumcount()

# 2) create entry 
prefix = "datasetR"   
df["entry"] = [f"{prefix}_{i:05d}" for i in range(1, len(df) + 1)]

# 3) column reorder
desired_order = [
    "entry",
    "frag_idx",
    "source_accession",
    "fragment_type",
    "is_fragment",
    "sequence",
    "fragment_length",
    "source_length"
]

df = df[desired_order]

# 4) save
df.to_csv("run_random_equal_val.csv", index=False)
```
