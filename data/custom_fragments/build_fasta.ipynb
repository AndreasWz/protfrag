```python
import pandas as pd
```

```python
df = pd.read_csv("dataset3_distribution_matched.csv")
df
```

```python
fasta = "\n".join(
    f"> id: {row.source_accession} type: {row.fragment_type} is_fragment: {row.is_fragment} count: {i}\n{row.sequence}"
    for i, row in enumerate(df.itertuples(index=False))
)

print(fasta.count("\n"))
```

```python
with open("dataset3.fasta", "w") as f:
    f.write(fasta)
```
